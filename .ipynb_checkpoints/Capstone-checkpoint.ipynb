**First of all, the FULL-TIME RESULT column (FTR) has two classifiers 'H' for home win and 'NH' that embraces both a draw and a away team victory. I want to be more specific hence I'm changing that to 'H' for home team win, 'D' for draw and 'A' for away team win.**

```python
new_ftr = []

for i in range(len(data)):
    if data['FTHG'][i] == data['FTAG'][i]:
        new_ftr.append('D')
    elif data['FTHG'][i] > data['FTAG'][i]:
        new_ftr.append('H')
    elif data['FTHG'][i] < data['FTAG'][i]:
        new_ftr.append('A')
        
data['FTR'] = new_ftr
```

```python
# Now I'm building a function that iterates trought the dataset and the derbies_dict to verify if both 
# Home and Away teams are included 

derbies_column = []

for row in range(len(data)):
    
    row_temp = []
    
    for team_1, team_2 in derbies_dict.items():
        
        if data['HomeTeam'][row] == team_1 and data['AwayTeam'][row] == team_2:
            row_temp.append(1)
        
        elif data['HomeTeam'][row] == team_2 and data['AwayTeam'][row] == team_1:
            row_temp.append(1)
        
        else:
            row_temp.append(0)
            
    if 1 in row_temp:
        derbies_column.append(1)
        
    else:
        derbies_column.append(0)

data['Derby'] = derbies_column
```

```python
# Let's start with the HTP column

new_htp = []

for row in range(len(data)):
    
    df_temp = data[data['Season'] == data['Season'][row]][data['Date'] < data['Date'][row]]
    
    team_points_temp = []
    
    for row2 in range(len(df_temp[df_temp['HomeTeam'] == data['HomeTeam'][row]])):
    
        if row2 > len(df_temp[df_temp['HomeTeam'] == data['HomeTeam'][row]]): 
        
            break
        
        else:
        
            if df_temp[df_temp['HomeTeam'] == data['HomeTeam'][row]]['FTR'].iloc[row2] == 'A':
                
                team_points_temp.append(0)
            
            elif df_temp[df_temp['HomeTeam'] == data['HomeTeam'][row]]['FTR'].iloc[row2] == 'D':
            
                team_points_temp.append(1)
               
            elif df_temp[df_temp['HomeTeam'] == data['HomeTeam'][row]]['FTR'].iloc[row2] == 'H':
               
                team_points_temp.append(3)

    for row3 in range(len(df_temp[df_temp['AwayTeam'] == data['HomeTeam'][row]])):
    
        if row3 > len(df_temp[df_temp['AwayTeam'] == data['HomeTeam'][row]]): 
        
            break
        
        else:
        
            if df_temp[df_temp['AwayTeam'] == data['HomeTeam'][row]]['FTR'].iloc[row3] == 'H':
                
                team_points_temp.append(0)
               
            elif df_temp[df_temp['AwayTeam'] == data['HomeTeam'][row]]['FTR'].iloc[row3] == 'D':
               
                team_points_temp.append(1)
               
            elif df_temp[df_temp['AwayTeam'] == data['HomeTeam'][row]]['FTR'].iloc[row3] == 'A':
               
                team_points_temp.append(3)
        
    new_htp.append(sum(team_points_temp))
```

**I'm going to build a new df having the different seasons as columns, the teams as rows and fill with the total points achieved by each team on that given season. With that information, I'll make another four columns on the main df, total championships won (HTCW and ATCW) and total points scored for each team in the previous season (HTLSP and ATLSP).**

**The first season (2000-2001) will be filled with information found on the web and the rest I'm able to do with the main df itself.** 

**All that engineering is been made to give to the model as much information as possible especially to help predict the first matches of the season where there's not much data to work with.**

```python
# First I'll create a dictionary with season 1999-2000 

total_points_1999_00 = {'manchester_united':91,
                        'arsenal':73,
                        'leeds':69,
                        'liverpool':67,
                        'chelsea':65,
                        'aston_villa':58,
                        'sunderland':58,
                        'leicester':55,
                        'west_ham':55,
                        'tottenham':53,
                        'newcastle':52,
                        'middlesbrough':52,
                        'everton':50,
                        'coventry':44,
                        'southampton':44,
                        'derby':38,
                        'bradford':36,
                        'wimbledon':33,
                        'sheffield_united':31,
                        'watford':24
                       }
```

```python
# Now I'm going to build a function to create the new df 

dict_total_points_per_season = {}

team_list = []

for season in data['Season'].unique():
    
    temp_season_list = []
        
    for team in data['HomeTeam'].unique():
        
        if team not in team_list:
            team_list.append(team)
        
        if season == '2000-2001' and team in total_points_1999_00:
            temp_season_list.append(total_points_1999_00[team])
            
        else:
            
            if team in data['HomeTeam'][data['MW'] == 38][data['Season']==season].unique():
                
                former_points = data['HTP2'][data['HomeTeam']== team][data['MW'] == 38][data['Season']==season].values[0]
                
                if data['FTR'][data['HomeTeam']== team][data['MW'] == 38][data['Season']==season].values[0] == 'A':
                    temp_season_list.append(former_points)
                    
                elif data['FTR'][data['HomeTeam']== team][data['MW'] == 38][data['Season']==season].values[0] == 'D':
                    temp_season_list.append((former_points+1))
                
                elif data['FTR'][data['HomeTeam']== team][data['MW'] == 38][data['Season']==season].values[0] == 'H':
                    temp_season_list.append((former_points+3))
            
            elif team in data['AwayTeam'][data['MW'] == 38][data['Season']==season].unique():
                
                former_points = data['ATP2'][data['AwayTeam']== team][data['MW'] == 38][data['Season']==season].values[0]
                
                if data['FTR'][data['AwayTeam']== team][data['MW'] == 38][data['Season']==season].values[0] == 'A':
                    temp_season_list.append((former_points+3))
                    
                elif data['FTR'][data['AwayTeam']== team][data['MW'] == 38][data['Season']==season].values[0] == 'D':
                    temp_season_list.append((former_points+1))
                
                elif data['FTR'][data['AwayTeam']== team][data['MW'] == 38][data['Season']==season].values[0] == 'H':
                    temp_season_list.append(former_points)
            
            else:
                temp_season_list.append(0)
                
                    
    dict_total_points_per_season[f'{season}'] = temp_season_list
                
dict_total_points_per_season['team'] = team_list
```

```python
df_total_points = pd.DataFrame.from_dict(dict_total_points_per_season)
```

```python
# Adding the season 1999-00 to that df as well

points_1999_2000 = []

for team in df_total_points['team']:
    if team in total_points_1999_00.keys():
        points_1999_2000.append(total_points_1999_00[team])
    else:
        points_1999_2000.append(0)

df_total_points['1999-2000'] = points_1999_2000
```

**Modelling**

I'll use all features to run the first model, just to see how it goes.

The result will be measured by 'accuracy' as we aim to predict the exact result of a match.

```python
# Before starting the modelling part, I have to define the baseline. 

data['FTR'].value_counts(normalize=True)
```

```python
# In this first model, I'm using the team names, despite presuming it won't help the model.

# First, let's pop it off the unnecessary columns. 

data_to_model_v1 = data.drop(columns=['Season', 'Date', 'MW', 'FTR', 'FTHG', 'FTAG'])
```

```python
# Dummifying categorical columns.

data_to_model_v1_dummy = pd.get_dummies(data_to_model_v1, drop_first=True)
data_to_model_v1_dummy
```

```python
# Instantiating predicted and predictors.

X_v1 = data_to_model_v1_dummy
y_v1 = data['FTR']
```

```python
# Performing a train-test split

X_train_v1, X_test_v1, y_train_v1, y_test_v1 = train_test_split(
    X_v1, y_v1, test_size=0.2, random_state=3, stratify=y_v1)
```

```python
# Rescaling the all dataset

scaler = StandardScaler()

X_train_v1_sst = pd.DataFrame(scaler.fit_transform(X_train_v1), columns=X_train_v1.columns)
X_test_v1_sst = pd.DataFrame(scaler.transform(X_test_v1), columns=X_test_v1.columns)
```

```python
# I'm gonna run a simple logistic regression with no tweaks

model_v1 = LogisticRegression(penalty='none', multi_class='ovr', max_iter=3000)
model_v1.fit(X_train_v1_sst, y_train_v1)
model_v1.score(X_test_v1_sst, y_test_v1)

print('Train score is:', model_v1.score(X_train_v1_sst, y_train_v1))
print('Test score is:', model_v1.score(X_test_v1_sst, y_test_v1))

scores_v1 = cross_val_score(model_v1, X_train_v1_sst, y_train_v1, cv=5)

print('CV scores are:', scores_v1)

print('Mean CV score is:', scores_v1.mean())
```

```python
predictions_test_v1 = model_v1.predict(X_test_v1_sst)

print(metrics.classification_report(y_test_v1, predictions_test_v1, zero_division=0))

metrics.plot_confusion_matrix(model_v1, X_test_v1_sst, y_test_v1, cmap='Blues')
plt.show()
```

Just above the baseline (as expected). I'm dropping the team names column and see if it improves the result.
